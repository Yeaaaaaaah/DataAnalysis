### 그리드 설정 (grid)


```python
# 보조선 활성화/비활성화 : plt.grid(T/F)
def draw_line(grid=False, style='b-'):
    plt.title('나이와 재산의 상관관계')
    plt.plot(
        range(10, 50, 10), # x축 데이터
        [1, 4, 9, 16], # y축 데이터
        style # 차트의 스타일
    )
    # plt.xlabel('x축 라벨')
    plt.xlabel('나이')
    plt.grid(grid)
    # plt.ylabel('y축 라벨')
    plt.ylabel('재산')
    plt.show()
```

```python
draw_line(grid=True) # 기본값은 False.
draw_line(grid=False)
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.subplot(2, 1, 1)
plt.plot(np.arange(5) ** 2)
plt.title("차트1")
plt.subplot(2, 1, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title("차트2")
plt.show() # 겹치는 문제가 발생함.
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.subplot(2, 1, 1)
plt.plot(np.arange(5) ** 2)
plt.title("차트1")
plt.subplot(2, 1, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title("차트2")
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show() 
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
plt.subplot(1, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title("차트1")
plt.subplot(1, 2, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title("차트2")
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show() 
```

```python
# 2x2 플롯 배치
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title("차트1")
plt.xlabel("x축")
plt.ylabel("y축")
plt.subplot(2, 2, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title("차트2")
plt.subplot(2, 2, 3)
plt.plot(np.arange(5) ** 3, 'g-')
plt.title("차트3")
plt.subplot(2, 2, 4)
plt.plot(np.arange(10), np.arange(10) % 3, 'b^--')
plt.title("차트4")
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show() 
```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y = [2,3,1]
x = np.arange(3)
xlabel = ["가", '나', '다'] #len(x) 길이와 같아야함
plt.title('바 차트')
plt.xticks(x, xlabel) #x와 x라벨이 같음 = x축의 개별 값 이름에다 이름을 붙여줌
plt.yticks(sorted(y)) #y 축에다가 기준이 되는 이름을 붙여주기 
plt.bar(x,y)
plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # len(x) 길이 같아야함
plt.title('가로 바 차트')
plt.yticks(x, xlabel) # x축의 개별 값에다가 이름 붙여줌
plt.xticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기
plt.barh(x, y) #barh = 가로 바차트. x와 y 뒤바뀜 주의
plt.show()
```

```python
people = ['james', 'john', 'kate', 'alice']
y_pos = np.arange(4)
performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar-h Chart')
plt.grid(True)
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x축')
plt.show()
```

#### 파이 차트
